```python
# Dependencies and Setup
import pandas as pd
import os
import csv
```

```python
# Set dataset path
orig_data_path = "OrigDataset/Wellbeing_and_lifestyle_data_Kaggle.csv"
```

```python
# Read in the dataset
orig_data = pd.read_csv(orig_data_path)

# Display 
orig_data.head()
```

```python
# Convert Timestamp column to datetime for consistent formatting
orig_data['Timestamp'] = pd.to_datetime(orig_data['Timestamp'])
```

```python
# Drop rows with timestamps prior to March 2019
clean_data = orig_data[orig_data['Timestamp']  >= '2019-03-01 00:00:00']

# Create new dataframe for the clean dataset
clean_data_df = pd.DataFrame(clean_data)
clean_data_df
```

```python
# Reassign Timestamp as the index to drop the row count column
clean_data_df.set_index('Timestamp', inplace=True)
clean_data_df
```

```python
# Save cleaned dataset to CSV 
clean_data_df.to_csv(r'CleanDataset\clean_data.csv')
```
